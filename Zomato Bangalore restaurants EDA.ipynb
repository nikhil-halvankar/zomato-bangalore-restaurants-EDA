```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
data =  pd.read_csv("zomato.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.columns
```

# dropping unnecessory columns

```python
data = data.drop(columns = ['url','address','phone','dish_liked','reviews_list','menu_item','listed_in(city)'], axis = 1)
```

```python
data.head()
```

```python
data.info()
```

# Droping duplicates values

```python
data.drop_duplicates(inplace = True)
data.shape
```

# Dealing with null values and cleaning columns 

```python
data.info()
```

```python
data['rate'].unique()         # there is values as NEW and - which is to be removed
```

```python
def handle_rate(i):
    if (i == 'NEW' or i == '-'):
        return np.nan
    else : 
        i = str(i).split('/')
        i = i[0]
        return float(i)
data['rate'] = data['rate'].apply(handle_rate)
data['rate'].unique()
```

```python
data['rate'].isnull().sum()
```

```python
data['rate'] = data['rate'].fillna(data['rate'].mean())          # fill null values with mean
data['rate'].isnull().sum()
```

```python
data.head()
```

```python
data['votes'].unique()   # votes col is ok
```

```python
data.info()
```

```python
data.dropna(inplace=True)         # other col have null values but in less qty. so droping it
data.info()
```

```python
data.head()
```

```python
data = data.rename(columns = {'approx_cost(for two people)':'cost_for_2', 'listed_in(type)':'type'})
data.head()
```

```python
data['cost_for_2'].unique()    # values in str ..making it float
```

```python
def handle_cost(j):
    if ',' in j:
        j = j.replace(',','')
        return float(j)
    else : 
        return float(j)
data['cost_for_2'] = data['cost_for_2'].apply(handle_cost)
data['cost_for_2'].unique()
```

```python
data['type'].unique()
```

```python
data['location'].unique()
```

```python
data['location'].value_counts()    # there are some min counts loaction
```

```python
loc = data['location'].value_counts(ascending = False)
loc_less_100 = loc[loc < 100]
loc_less_100
```

```python
def handle_location(k):
    if (k in loc_less_100):
        return 'others'
    else :
        return k
data['location'] = data['location'].apply(handle_location)
data['location'].value_counts()
```

```python
data.head()
```

```python
data['rest_type'].value_counts()           # there are some min counts rest type
```

```python
rest = data['rest_type'].value_counts(ascending = False)
rest_less_100 = rest[rest < 100]
rest_less_100
```

```python
def handle_rest(m):
    if(m in rest_less_100):
        return 'others'
    else :
        return m
data['rest_type'] = data['rest_type'].apply(handle_rest)
data['rest_type'].value_counts()
```

```python
data.head()
```

```python
data['cuisines'].value_counts()
```

```python
cu = data['cuisines'].value_counts(ascending = False)
cu_less_100 = cu[cu < 100]
cu_less_100
```

```python
def handle_cu(n):
    if(n in cu_less_100):
        return 'others'
    else : 
        return n
data['cuisines'] = data['cuisines'].apply(handle_cu)
data['cuisines'].value_counts()
```

```python
data.head()
```

```python
# countplot for location

plt.figure(figsize = (15,7))
sns.countplot(data['location'])
plt.xticks(rotation=90)
plt.show()
```

```python
# Relation bet online order vs ratings

plt.figure(figsize = (7,6))
sns.boxplot('online_order','rate',data=data)
plt.show()
```

```python
# Relation bet book table vs ratings

plt.figure(figsize = (7,6))
sns.boxplot('book_table','rate',data=data)
plt.show()
```

```python
# online order facility location wise

df1 = data.groupby(['location','online_order'])['name'].count()
df1.to_csv('location_online')
df1 = pd.read_csv("location_online")
df1 = df1.pivot_table(df1, index = 'location', columns = 'online_order')
df1.head()
```

```python
df1.plot(kind='bar',figsize = (18,6))
plt.show()
```

```python
# book table facility location wise

df2 = data.groupby(['location','book_table'])['name'].count()
df2.to_csv('location_table')
df2 = pd.read_csv("location_table")
df2 = df2.pivot_table(df2, index = 'location', columns = 'book_table')
df2.head()
```

```python
df2.plot(kind='bar',figsize = (18,6))
plt.show()
```

```python
# Relation bet types vs ratings

plt.figure(figsize= (15,6))
sns.boxplot('type','rate',data=data)
plt.show()
```

```python
#  type location wise

df3 = data.groupby(['location','type'])['name'].count()
df3.to_csv('location_type')
df3 = pd.read_csv("location_type")
df3 = df3.pivot_table(df3, index = 'location', columns = 'type')
df3.head()
```

```python
df3.plot(kind='bar',figsize = (18,6))
plt.show()
```

```python
# Relation bet votes vs location

df4 = data[['location','votes']]
df4.drop_duplicates()
df5 = df4.groupby(['location'])['votes'].sum()
df5 = df5.to_frame()
df5 = df5.sort_values('votes',ascending = False)
df5.head()
```

```python
df5.plot(kind='bar',figsize = (18,6))
plt.show()
```

```python
# Relation bet cuisines vs votes

df6 = data[['cuisines','votes']]
df6.drop_duplicates()
df7 = df6.groupby(['cuisines'])['votes'].sum()
df7 = df7.to_frame()
df7 = df7.sort_values('votes',ascending = False)
df7 = df7.iloc[1:,:]
df7.head()
```

```python
df7.plot(kind='bar',figsize = (18,6))
plt.show()
```
